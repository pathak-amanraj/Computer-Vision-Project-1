```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os

# -------------------------------------------------
# 1. Load Dataset Using image_dataset_from_directory
# -------------------------------------------------
# Path to your final processed dataset (each class in its own subfolder)
data_dir = "/kaggle/working/final_processed_dataset"

# Parameters
batch_size = 32
img_size = (224, 224)
seed = 123

# Create training and validation datasets (80-20 split)
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=seed,
    image_size=img_size,
    batch_size=batch_size
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=seed,
    image_size=img_size,
    batch_size=batch_size
)

# Save class names and number of classes before applying prefetch
class_names = train_ds.class_names
num_classes = len(class_names)
print("Class names:", class_names)

# -------------------------------------------------
# 2. Create an On-the-Fly Data Augmentation Layer
# -------------------------------------------------
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
])

# Prefetch for performance
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)

# -------------------------------------------------
# 3. Build the ResNet-Based Model with Transfer Learning
# -------------------------------------------------
# Define the input layer
inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))

# Apply data augmentation
x = data_augmentation(inputs)

# Preprocess inputs in the same way as ResNet50 expects
x = tf.keras.applications.resnet50.preprocess_input(x)

# Load the ResNet50 base model (with ImageNet weights, excluding the top)
base_model = tf.keras.applications.ResNet50(
    weights='imagenet', 
    include_top=False, 
    input_tensor=x
)

# Initially freeze the base model
base_model.trainable = False

# Add custom top layers
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(0.5)(x)

# Final classification layer: use the number of classes from the variable we saved
outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(x)

# Create the model
model = tf.keras.Model(inputs, outputs)

# Compile the model with Adam optimizer and an initial learning rate
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# -------------------------------------------------
# 4. Define Callbacks to Prevent Overfitting
# -------------------------------------------------
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', 
    patience=5, 
    restore_best_weights=True, 
    verbose=1
)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', 
    factor=0.5, 
    patience=3, 
    verbose=1
)
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    'best_resnet_model.keras', 
    monitor='val_loss', 
    save_best_only=True, 
    verbose=1
)
callbacks = [early_stop, reduce_lr, checkpoint]

# -------------------------------------------------
# 5. Train the Model (Initial Phase with Base Frozen)
# -------------------------------------------------
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30,
    callbacks=callbacks
)

# -------------------------------------------------
# 6. Fine-Tune the Model by Unfreezing Part of the Base
# -------------------------------------------------
# Unfreeze the base model for fine-tuning
base_model.trainable = True
# Optionally freeze the earlier layers (for example, freeze first 100 layers)
for layer in base_model.layers[:100]:
    layer.trainable = False

# Re-compile with a lower learning rate for fine-tuning
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_ft = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=callbacks
)

# -------------------------------------------------
# 7. Evaluate and Visualize the Results
# -------------------------------------------------
# Plot training and validation accuracy
plt.figure(figsize=(10, 5))
# Concatenate history from both training phases
train_acc = history.history['accuracy'] + history_ft.history['accuracy']
val_acc = history.history['val_accuracy'] + history_ft.history['val_accuracy']
plt.plot(train_acc, label='Train Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy (ResNet with Fine-Tuning)')
plt.show()

# Evaluate on the validation set
val_loss, val_accuracy = model.evaluate(val_ds)
print("Validation Accuracy: {:.4f}".format(val_accuracy))
```

```python
# 8. Evaluate on Test Dataset
# -------------------------------------------------
# Assume you have a separate test dataset folder organized similarly
test_data_dir = "/kaggle/working/final_processed_dataset"  # Adjust this path as needed

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    test_data_dir,
    image_size=img_size,
    batch_size=batch_size,
    shuffle=False
)
test_ds = test_ds.prefetch(buffer_size=AUTOTUNE)

test_loss, test_accuracy = model.evaluate(test_ds)
print("Test Accuracy: {:.4f}".format(test_accuracy))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score
from tensorflow.keras.utils import to_categorical

# --------------------------------------------
# 1. Extract true labels from the test dataset
# --------------------------------------------
# Here, we assume that test_ds is a tf.data.Dataset returning (images, labels) batches.
# We concatenate all true labels into one numpy array.
y_true = np.concatenate([y for _, y in test_ds], axis=0)

# Convert true labels to one-hot encoding
y_true_onehot = to_categorical(y_true, num_classes=num_classes)

# -----------------------------------------------------
# 2. Obtain predicted probabilities from the test dataset
# -----------------------------------------------------
y_probs = model.predict(test_ds)

# -----------------------------------------------------
# 3. Compute and Plot Precision-Recall Curves for Each Class
# -----------------------------------------------------
plt.figure(figsize=(10, 6))
for i in range(num_classes):
    precision, recall, thresholds = precision_recall_curve(y_true_onehot[:, i], y_probs[:, i])
    plt.plot(recall, precision, marker='.', label=f'PR Curve for class: {class_names[i]}')
    
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves for Each Class')
plt.legend()
plt.show()

# -----------------------------------------------------
# 4. Compute and Print the Average Precision Score for Each Class
# -----------------------------------------------------
for i in range(num_classes):
    ap = average_precision_score(y_true_onehot[:, i], y_probs[:, i])
    print(f"Average Precision for class '{class_names[i]}': {ap:.2f}")
```
